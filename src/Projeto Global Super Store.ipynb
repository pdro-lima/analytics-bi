```python
import pandas as pd
```

```python
# Lendo base de dados de itens de pedidos

base_itens_pedidos = pd.read_excel('..\data\Global Superstore.xls', sheet_name=0)
```

```python
#Tratando Customer ID diferentes pra mesma pessoa

def corrige_customer_id(id):
    if len(id) < 8:
        elementos_id = id.split('-')
        elementos_id[1] = "1" + elementos_id[1]
        id = '-'.join(elementos_id)
        return id
    else:
        return id


base_itens_pedidos['Customer ID'] = base_itens_pedidos['Customer ID'].apply(corrige_customer_id)
```

```python
# Criando Base de clientes
base_clientes = pd.DataFrame()

base_clientes['Customer ID'] = base_itens_pedidos['Customer ID'].unique()

base_clientes = base_clientes.merge(
    base_itens_pedidos[['Customer ID', 'Customer Name', 'Segment']],
    on='Customer ID',
    how='left'
).drop_duplicates(subset='Customer ID')
```

```python
# Criando Base de produtos
base_produtos = pd.DataFrame()

base_produtos['Product ID'] = base_itens_pedidos['Product ID'].unique()

base_produtos = base_produtos.merge(
    base_itens_pedidos[['Product ID', 'Product Name', 'Category', 'Sub-Category']],
    on='Product ID',
    how='left'
).drop_duplicates(subset='Product ID')
```

```python
# Fazendo a normalização da tabela fato itens_pedidos

base_itens_pedidos = base_itens_pedidos[['Row ID', 
                                         'Order ID', 
                                         'Order Date', 
                                         'Ship Date', 
                                         'Ship Mode', 
                                         'Customer ID', 
                                         'City', 'State', 
                                         'Country', 
                                         'Postal Code', 
                                         'Market', 
                                         'Region', 
                                         'Product ID', 
                                         'Sales', 
                                         'Quantity', 
                                         'Discount', 
                                         'Profit', 
                                         'Shipping Cost', 
                                         'Order Priority']]
```

```python
# Salvando as tabelas

base_itens_pedidos.to_parquet('..\data\itens_pedidos.gzip', index=False, compression='gzip')
base_produtos.to_excel('..\data\produtos.xlsx', index=False)
base_clientes.to_excel('..\data\clientes.xlsx', index=False)
```
